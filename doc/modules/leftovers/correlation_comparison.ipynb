```python
### Felekyan
#  Read the data data
data = tttrlib.TTTR('./tttr-data/bh/bh_spc132.spc', 'SPC-130')

import tttrlib
import matplotlib.pylab as plt
import numpy as np
import pathlib

ch1 = [8, ]
ch2 = [0, ]

# Correlator settings, if the same settings are used repeatedly it is useful to define them once
settings = {
    "method": "default",
    "n_bins": 9,  # n_bins and n_casc defines the settings of the multi-tau
    "n_casc": 25,  # correlation algorithm
    "make_fine": False  # Do not use the microtime information
}

# Create correlator
# Caution: x-axis in units of macro time counter
# tttrlib.Correlator is unaware of the calibration in the TTTR object
correlator = tttrlib.Correlator(**settings)
# Select the green channels (channel number 0 and 8)
ch1_indices = data.get_selection_by_channel(ch1)
ch2_indices = data.get_selection_by_channel(ch2)

# Select macro times for Ch1 and Ch2 and create array of weights
# Note: the weights are set to 1, i.e. no weighting is used
t = data.macro_times
t1 = t[ch1_indices]
w1 = np.ones_like(t1, dtype=np.float64)
t2 = t[ch2_indices]
w2 = np.ones_like(t2, dtype=np.float64)
correlator.set_events(t1, w1, t2, w2)


x1 = correlator.curve.x
y1 = correlator.get_corr_normalized()

plt.plot(y1, label="Gp/Gs")
```

```python
settings = {
    "method": "laurence",
    "n_bins": 9,  # n_bins and n_casc defines the settings of the multi-tau
    "n_casc": 25,  # correlation algorithm
    "make_fine": False  # Do not use the microtime information
}

# Create correlator
# Caution: x-axis in units of macro time counter
# tttrlib.Correlator is unaware of the calibration in the TTTR object
correlator = tttrlib.Correlator(**settings)

# Select the green channels (channel number 0 and 8)
ch1_indices = data.get_selection_by_channel(ch1)
ch2_indices = data.get_selection_by_channel(ch2)

t = data.macro_times
w1 = np.zeros_like(t, dtype=np.float64)
w2 = np.zeros_like(t, dtype=np.float64)
w1[ch1_indices] = 1.0
w2[ch2_indices] = 1.0
correlator.set_events(t, w1, t, w2)

x2 = correlator.curve.x
y2 = correlator.get_corr_normalized()

plt.semilogy(y2, label="Gp/Gs")
#plt.ylim(0, 20)

plt.plot(y2, label="Laurence")
plt.plot(y1, label="Fast")
plt.legend()
plt.ylim(0, 20)

plt.plot(y2[:-1] - y1[:-1])
plt.ylim(-0.2, 0.2)
```
